## OLS 메소드를 사용한 단순 선형 회귀

- OLS:잔차제곱합(SSR)을 최소화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips)
```

```python
import statsmodels.api as sm

lr = sm.OLS.from_formula("tip ~ total_bill", data = tips).fit()

print(lr.summary())
```

```python
# 다중 선형 회귀

from sklearn import datasets

boston = datasets.load_boston()

boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
print(boston.DESCR)
```

```python
boston_df.head()
```

```python
boston_df["target"] = boston.target

boston_df.head()
```

```python
import statsmodels.formula.api as smf

model = smf.ols(formula = "target ~ CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B+LSTAT", data = boston_df)
result = model.fit()

print(result.summary())
```

```python
# 다중공선성 존재 -> 해결법 찾아보기

model = smf.ols(formula = "target ~ CRIM+ZN+INDUS+CHAS+NOX+RM+DIS+RAD+TAX+PTRATIO+B+LSTAT", data = boston_df)
result = model.fit()

print(result.summary())
```

```python
# sklearn 으로 회귀모델 생성

from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

boston = load_boston()

y_data = boston.target
X_data = boston.data

print(X_data.shape)
print(y_data.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = .3)

lr = LinearRegression()
lr.fit(X_train, y_train)

y_predict = lr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)

print("MSE : {0:.3f}, RMSE : {1:.3f}".format(mse, rmse))
print("R^2 : {0:.3f}".format(r2_score(y_test, y_predict)))
print("lr_score : {0:.3f}".format(lr.score(X_test, y_test)))
```

```python
print("회귀계수 : ", np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data = np.round(lr.coef_, 1), index = boston.feature_names)
coef
```

```python
fig, axes = plt.subplots(figsize = (16,16), ncols = 4, nrows = 4)

for i, feature in enumerate(boston.feature_names):
    row = int(i / 4)
    col = i % 4
    sns.regplot(x = feature, y = "target", data = boston_df, ax = axes[row][col])
```

```python
# Logistic Regression

from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer(as_frame = True)

print(cancer.DESCR)
```

```python
cancer_df = cancer.frame

cancer_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

cancer_scaled = scaler.fit_transform(cancer.data)

print(cancer_scaled[0])
```

```python
from sklearn.linear_model import LogisticRegression

y = cancer_df["target"]
X = cancer_scaled

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

confusion_matrix(y_test, y_predict)
```

```python
accuracy = accuracy_score(y_test, y_predict)

print(accuracy.round(3))
```
