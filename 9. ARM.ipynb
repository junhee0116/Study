```python
# apriori 알고리즘
from mlxtend.frequent_patterns import apriori, association_rules

# 연관분석용 데이터 파일 변환
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tran = [["bread", "milk"],
        ["bread", "diaper", "beer", "eggs"],
        ["milk", "diaper", "beer", "coke"],
        ["bread", "milk", "diaper", "beer"],
        ["bread", "milk", "diaper", "coke"]]
```

```python
# 데이터 세트를 연관분석이 가능한 배열로 변환 
te = TransactionEncoder()

# fit 함수로 데이터셋은 고유 레이블을 갖고
tran_ar = te.fit(tran).transform(tran)
print(tran_ar)
```

```python
# 열 이름
print(te.columns_)
```

```python
# 각 상품명을 열이름으로 하는 데이터 프레임으로 변환
df = pd.DataFrame(tran_ar, columns = te.columns_)
print(df)
```

```python
# 각 상품별 거래 빈도
freq = df.sum().to_frame("Frequency")

# 빈도 역순으로 정렬
freq_sort = freq.sort_values("Frequency", ascending = False)
print(freq_sort)
```

```python
freq_sort["Support"] = freq_sort["Frequency"] / len(df)
print(freq_sort)
```

```python
freq_sort["Frequency"].plot(kind = "bar", title = "FrequentItems")
plt.show()
```

```python
freq_sort["Support"].hist(bins = 6)
plt.show()
```

```python
# 최소지지도 0.4 빈발항목집합
freq_items = apriori(df, min_support = 0.4, use_colnames = True)
print(freq_items)
```

```python
rules = association_rules(freq_items, metric = "confidence", min_threshold = 0.7)
rules
```

```python
# lift가 1보다 큰 양의 관계를 만족한느 연관규칙들을 추출하고
rules = rules[(rules["lift"] > 1)]

# 신뢰도의 내림차순으로 정렬
rules.sort_values(by = "confidence", ascending = False)
```

```python
# 조건부에 "diaper"가 있는 규칙
rules[rules.antecedents.apply(lambda x: "diaper" in x)]
```

```python
# 결론부에 "milk"가 있는 규칙
rules[rules.consequents.apply(lambda x: "milk" in x)]
```

```python
df = pd.read_excel("Online Retail.xlsx")
df
```

```python
df = df[["InvoiceNo", "Description", "Quantity", "UnitPrice"]]
```

```python
df[df["Quantity"] <= 0]
```

```python
# 주문량이 0보다 큰 거래만 추출
df = df[df["Quantity"] > 0]
df
```

```python
# 거래별 상품명별 수량 합계
df = df.groupby(["InvoiceNo", "Description"])["Quantity"].sum()
df
```

```python
# 데이터프레임의 구조변경
# 1열 InvoiceNo는 행으로, 2열 Description은 열로
df= df.unstack()
df
```

```python
# 각 셀의 수량이 1 이상이면 1로, NaN이면 0으로 변환
## 함수 : 전달되는 x값들을 조건에 따라 0, 1로 변환
def encode_units(x):
    if x >= 1:
        return 1
    else:
        return 0
    
# df 데이터세트에 대해 encode_units 함수 적용
df = df.applymap(encode_units)
df
```

```python
# 각 열별로 합치고 내림차순으로 정렬하여 가장 많은 수량으로 거래된 상품명 확인하기
# 아이템별 빈도 및 항목명 설정
freq = df.sum(axis = 0).to_frame("Frequency")

# 빈도의 내림차순 정렬
freq_sort = freq.sort_values("Frequency", ascending = False)

freq_sort
```

```python
# 빈도수가 1인 항목 제외
freq_sort = freq_sort[freq_sort["Frequency"] > 1]
freq_sort
```

```python
freq_sort["Support"] = freq_sort["Frequency"] / len(freq_sort)
freq_sort
```
